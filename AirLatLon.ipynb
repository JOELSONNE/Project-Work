```python
import pandas as pd

air_pre = pd.read_csv('/Users/jes/Library/CloudStorage/OneDrive-RobertGordonUniversity(2)/ProjectMINE/Birmingham/Air_Pre_Clean.csv')
air_post = pd.read_csv('/Users/jes/Library/CloudStorage/OneDrive-RobertGordonUniversity(2)/ProjectMINE/Birmingham/Air_Post_Clean.csv')
print(air_pre['site_name'].unique())
print(air_post['site_name'].unique())
```

```python
def normalize_name(x):
    x= str(x ).lower()
    if '4540' in x: return 'Birmingham A4540 Roadside'
    if 'acock' in x: return 'Birmingham Acocks Green'
    if 'hall' in x: return 'Hall Green'
    if 'lady' in x: return 'Birmingham Ladywood'
    if 'oldbury' in x: return 'Oldbury Birmingham Road'
    return x
    
air_pre['site_name'] = air_pre['site_name'].apply(normalize_name)
air_post['site_name'] = air_post['site_name'].apply(normalize_name)

print(air_pre['site_name'].unique())
print(air_post['site_name'].unique())
```

```python
air_sites = pd.DataFrame({
    'site_name': [
        'Birmingham A4540 Roadside',
        'Birmingham Acocks Green', 
        'Hall Green', 
        'Birmingham Ladywood', 
        'Oldbury Birmingham Road'
    ],
    'latitude': [
        52.476145, 
        52.437165, 
        52.437553, 
        52.481346, 
        52.502436
    ],
    'longitude': [
        -1.874978, 
        -1.829999, 
        -1.830956, 
        -1.918235, 
        -2.003497
   ]
})

air_pre = air_pre.merge(air_sites, on='site_name', how='left')
air_post = air_post.merge(air_sites, on='site_name', how='left')

air_pre.to_csv('/Users/jes/Library/CloudStorage/OneDrive-RobertGordonUniversity(2)/ProjectMINE/Birmingham/Air_Pre_Final.csv', index=False)
air_post.to_csv('/Users/jes/Library/CloudStorage/OneDrive-RobertGordonUniversity(2)/ProjectMINE/Birmingham/Air_Post_Final.csv', index=False)

print(air_pre[['site_name','latitude','longitude']].drop_duplicates())
print(air_post[['site_name','latitude','longitude']].drop_duplicates())
```
