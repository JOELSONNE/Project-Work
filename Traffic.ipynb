```python
import pandas as pd

# Load data
traffic = pd.read_csv('/Users/jes/Library/CloudStorage/OneDrive-RobertGordonUniversity(2)/ProjectMINE/Birmingham/Traffic_data.csv', low_memory=False)

# Convert count_date to proper datetime using day-first format (UK style)
traffic['count_date'] = pd.to_datetime(traffic['count_date'], errors='coerce', dayfirst=True)

# Drop rows where date or hour are missing
traffic = traffic.dropna(subset=['count_date', 'hour'])

# Ensure hour is integer
traffic['hour'] = traffic['hour'].astype(int)

# Combine into a single datetime column
traffic['datetime'] = traffic['count_date'] + pd.to_timedelta(traffic['hour'], unit='h')

# Drop duplicates based on unique site & timestamp
traffic = traffic.drop_duplicates(subset=['count_point_id', 'datetime'])

start_date = pd.Timestamp('2018-01-01')
traffic = traffic[traffic['datetime'] >= start_date].copy()

# Split pre/post CAZ (1 June 2021)

caz_date = pd.Timestamp('2021-06-01')
traffic_pre  = traffic[traffic['datetime'] <  caz_date].copy()
traffic_post = traffic[traffic['datetime'] >= caz_date].copy()

# Save both splits
traffic_pre.to_csv("/Users/jes/Library/CloudStorage/OneDrive-RobertGordonUniversity(2)/ProjectMINE/Birmingham/Traffic_Pre.csv", index=False)
traffic_post.to_csv("/Users/jes/Library/CloudStorage/OneDrive-RobertGordonUniversity(2)/ProjectMINE/Birmingham/Traffic_Post.csv", index=False)

print("✅ Traffic data split completed.")
print("Pre records:", len(traffic_pre), "| Post records:", len(traffic_post))
```
